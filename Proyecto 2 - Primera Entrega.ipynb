```python
electricity = wb.data.DataFrame('EG.ELC.ACCS.ZS', economy = ['ARG','BRA','VEN','BOL','CHL','ECU','PER','URY','PRY','COL','MEX','CRI','CAN','USA'], time=range(2002,2020), labels=True).reset_index(drop=True).sort_values(by='YR2016', ascending=False).set_index('Country')
electricity
```

```python
electricity = electricity.unstack().reset_index()
```

```python
population = wb.data.DataFrame('EN.POP.DNST', economy = ['ARG','BRA','VEN','BOL','CHL','ECU','PER','URY','PRY','COL','MEX','CRI','CAN','USA'], time=range(2002,2020), labels=True).reset_index(drop=True).sort_values(by='YR2016', ascending=False).set_index('Country')
population
```

```python
population = population.unstack().reset_index()
```

```python
hunger = wb.data.DataFrame('SN.ITK.DEFC.ZS', economy = ['ARG','BRA','VEN','BOL','CHL','ECU','PER','URY','PRY','COL','MEX','CRI','CAN','USA'], time=range(2002,2020), labels=True).reset_index(drop=True).sort_values(by='YR2016', ascending=False).set_index('Country')
hunger
```

```python
hunger = hunger.unstack().reset_index()
```

```python
unemployment = wb.data.DataFrame('SL.UEM.TOTL.ZS', economy = ['ARG','BRA','VEN','BOL','CHL','ECU','PER','URY','PRY','COL','MEX','CRI','CAN','USA'], time=range(2002,2020), labels=True).reset_index(drop=True).sort_values(by='YR2016', ascending=False).set_index('Country')
hunger
```

```python
unemployment = unemployment.unstack().reset_index()
```

```python
cultivate = wb.data.DataFrame('AG.LND.ARBL.ZS', economy = ['ARG','BRA','VEN','BOL','CHL','ECU','PER','URY','PRY','COL','MEX','CRI','CAN','USA'], time=range(2002,2020), labels=True).reset_index(drop=True).sort_values(by='YR2016', ascending=False).set_index('Country')
cultivate
```

```python
cultivate = cultivate.unstack().reset_index()
```

```python
bienestar = electricity.merge(population, on =['level_0', 'Country'],how='outer').merge(cultivate, on =['level_0', 'Country'],how='outer').merge(unemployment, on =['level_0', 'Country'],how='outer').merge(hunger, on =['level_0', 'Country'],how='outer')
```
